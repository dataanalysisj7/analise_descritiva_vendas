# Análise descritiva de vendas

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("chadwambles/supermarket-sales")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
df_data = pd.read_csv(path + "/sales.csv")
```

```python
df_data
```

## Qual o faturamento por filial?

```python
df_data.groupby("city")["total_price"].sum()
```

## Qual o percentual de participação na receita de cada tipo de produto?

```python
(df_data.groupby("product_category")["total_price"].sum() / df_data.groupby("product_category")["total_price"].sum().sum()).sort_values().round(2) * 100
```

## Como está distribuído o tipo de produto por gênero?

```python
df_data.groupby(["gender", "product_category"])[["total_price"]].sum().pivot_table(index = "product_category", columns="gender")
```

## Como se distribui o gasto por tipo de consumidor em cada filial?

```python
df_data.groupby(["customer_type", "city"])[["total_price"]].sum().round(2)
```
